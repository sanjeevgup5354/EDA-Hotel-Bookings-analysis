## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Delete the duplicates
df.drop_duplicates(inplace=True)

#deleting rows with null values
df.dropna(subset='country', inplace=True,axis=0)
df.dropna(subset='children', inplace=True,axis=0)

#replacing null values
df['agent'].fillna(df['agent'].median(),inplace=True)
df['company'].fillna(0, inplace=True)
````

````python
#Changing data types of columns
df['children']=df['children'].astype(int)
````

````python
#checking for outliers
sns.boxplot(data=df)
plt.xlabel("Column_name")
plt.title('Boxplot for Hotel Bookings Data')
plt.xticks(rotation=90)
plt.figure(figsize=(32, 12))
````

````python
# Remove outliers from adr column
df.drop(df[df['adr']>3000].index,inplace=True,axis=0)
````

### What all manipulations have you done and insights you found?

After analysing the data we found multiple duplicate rows as well as null values which wew repalced by their best fits.

Manipulations:
1. Deleted the duplicate rows.
2. Dropped the rows where country column was empty
3. Dropped the rows where children column was empty
4. Replaced null values in agent column with the median value of the entire coulmn
5. Replaced company coulmn null values with zero.
6.Checked for an outlier and found an outlier, removed it from the data.

#### Chart - 2 Distribution channel used for hotel bookings

````python
group_by_dc = df.groupby('distribution_channel')
df2 = pd.DataFrame(round(group_by_dc['lead_time'].median(),2)).reset_index().rename(columns = {'lead_time': 'median_lead_time'})
plt.figure(figsize = (7,5))
sns.barplot(x = df2['distribution_channel'], y = df2['median_lead_time'])
plt.show()
````

##### 1. Why did you pick the specific chart?

In Bar charts, individual bars or columns are used to represent different categories or groups. The purpose of using a bar graph is to visually display and compare the values of different variables or categories. In this case we have to find out the channel which is mostly used for hotel bookings which can be best shown using bar charts.
